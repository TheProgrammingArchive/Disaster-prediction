```python
# Import modules required
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read train data
train_data = pd.read_csv('/content/train.csv')
train_data
```

```python
# Removing irrelevant columns
train_data = train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Remove null
train_data = train_data.dropna()
train_data = train_data.reset_index()

train_data
```

```python
# Splitting training data
X_train = train_data.drop('Survived', axis=1)
y_train = train_data['Survived']
```

```python
X_train, y_train
```

```python
# Normalizing data
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

ct = make_column_transformer(
    (OneHotEncoder(), ['Sex', 'Embarked', 'Pclass']),
    (MinMaxScaler(), ['Age', 'Fare'])
)

ct.fit(X_train)

X_train_ = ct.transform(X_train)
```

```python
# Building a model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100),
  tf.keras.layers.Dense(50),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(1)
])

# Compiling
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(lr=0.0001),
              metrics=['mae'])

# Fit
model.fit(X_train_, y_train, epochs=200)
```

```python
# Gather test data
test_data = pd.read_csv('/content/test (1).csv')
test_id = test_data['PassengerId']
test_data = test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Drop null
test_data = test_data.dropna()
test_data = test_data.reset_index()
test_data
```

```python
# Normalize test
X_test_ = ct.transform(test_data)
```

```python
# Evaluate
y_preds = model.predict(X_test_)
y_preds.shape
```

```python
y_preds_list = []
y_pred_final = []

for pred in y_preds:
  y_preds_list.append(float(pred))

for x in y_preds_list:
  y_pred_final.append(round(x))
```
